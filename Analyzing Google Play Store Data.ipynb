```python
google_data.dropna(thresh=threshold,axis=1,inplace=True)
#1 for column
#0 for row
```

# Correct Data Types

```python
# Filling missing values for 'Rating' and 'Size' with mean (or median)
google_data['Rating'] = google_data['Rating'].fillna(google_data['Rating'].mean())  # Or use median
google_data['Size'] = google_data['Size'].fillna(google_data['Size'].mean())  # Or use median

# Filling missing values for categorical columns using mode
google_data['Current Ver'] = google_data['Current Ver'].fillna(google_data['Current Ver'].mode()[0])
google_data['Android Ver'] = google_data['Android Ver'].fillna(google_data['Android Ver'].mode()[0])


# Cleaning the Installs column
# Cleaning the Installs column
google_data['Installs'] = (  google_data['Installs'] .str.replace(',', '')   .str.replace('+', '', regex=False)  # Remove '+' symbols  .astype(int)                 # Convert to integers
)

# Convert the Installs column to numeric
google_data['Installs'] = pd.to_numeric(google_data['Installs'], errors='coerce')


```

# data visualization

```python
grp = google_data.groupby('Category')

# Use strings for the aggregation methods
x = grp['Installs'].agg("mean")
y = grp['Price'].agg("sum")
z = grp['Reviews'].agg("mean")

print(x)
print(y)
print(z)
```

```python
# Group data by category and calculate the sum of installs for each category
category_installs = google_data.groupby('Category')['Installs'].sum()

# Plotting the scatter plot
plt.figure(figsize=(16, 6))
plt.scatter(category_installs.index, category_installs, color='b', marker='o')
plt.xticks(rotation=90)
plt.title('Category vs Total Installs')
plt.xlabel('Categories')
plt.ylabel('Total Installs')
plt.show()
```

```python
# Count the number of reviews per category
category_reviews = google_data.groupby('Category')['Reviews'].sum().sort_values(ascending=False)

# Plotting a bar plot
plt.figure(figsize=(16, 6))
category_reviews.plot(kind='bar', color='skyblue')
plt.title('Total Reviews per Category')
plt.xlabel('Category')
plt.ylabel('Total Reviews')
plt.xticks(rotation=90)
plt.show()
```

```python
# Distribution of app ratings
plt.figure(figsize=(10, 6))
sns.histplot(google_data['Rating'], kde=True, color='blue', bins=20)
plt.title('Distribution of Ratings')
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate the sum of apps in each category
category_count=google_data['Category'].value_counts()
print('Number of app in each category')
print(category_count)
```

```python
# # Explore app distribution across categories
category_counts = google_data['Category'].value_counts()

# Plot the category distribution
plt.figure(figsize=(12, 6))
sns.barplot(x=category_counts.values, y=category_counts.index, color='teal')  # Use a single color
plt.title('Distribution of Apps by Category')
plt.xlabel('Number of Apps')
plt.ylabel('Category')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter apps with ratings >= 5
apps_with_high_rating = google_data[google_data['Rating'] >= 5]

# Sort the apps by Rating for better visualization and select the top 10
top_10_apps = apps_with_high_rating.sort_values(by='Rating', ascending=False).head(10)

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(
    y=top_10_apps['App'], 
    x=top_10_apps['Rating']
)
plt.title('Top 10 Apps with Ratings 5')
plt.xlabel('Rating')
plt.ylabel('App Name')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter apps with Rating == 5
apps_with_rating_5 = google_data[google_data['Rating'] == 5]

# Sort by App name for better visualization and limit to the first 10
top_10_apps = apps_with_rating_5[['App', 'Category', 'Rating']].head(20)

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(
    y=top_10_apps['App'], 
    x=top_10_apps['Rating'], 
    hue=top_10_apps['Category'], 
    dodge=False
)
plt.title('Top 20 Apps with Rating of 5 and Their Categories')
plt.xlabel('Rating')
plt.ylabel('App Name')
plt.legend(title='Category')
plt.show()
```

```python


# Group by Category and sum the installs
category_installs = google_data.groupby('Category')['Installs'].sum().sort_values(ascending=False)

# Find the category with the highest installs
highest_install_category = category_installs.idxmax()
highest_install_count = category_installs.max()

print(f"Category with the highest installs: {highest_install_category}")
print(f"Number of installs: {highest_install_count}")

# Visualization
plt.figure(figsize=(12, 6))
sns.barplot(x=category_installs.values, y=category_installs.index, palette='coolwarm')
plt.title('Total Installs by Category')
plt.xlabel('Total Installs')
plt.ylabel('Category')
plt.show()
```
